# **Statistics Advance Part 1**

1. What is a random variable in probability theory?
   -  A random variable is a function that assigns a numerical value to each outcome of a random experiment.

     - Example: Tossing a coin → Head = 1, Tail = 0.

2. What are the types of random variables?
    -  Discrete Random Variable: Takes countable values (e.g., number of heads in 3 coin tosses).

    - Continuous Random Variable: Takes any value in a range (e.g., temperature, height).
3. What is the difference between discrete and continuous distributions?
   -  Feature	: Values
  -  Discrete Distribution	: Countable
   - Continuous Distribution : Infinite/Uncountable

  -  Feature	: Example
   - Discrete Distribution : 	Binomial, Poisson
 -  Continuous Distribution : Normal, Uniform

  -  Feature	: Probability
  -  Discrete Distribution	: Exact values (P(X = x))
  -  Continuous Distribution : Ranges (P(a < X < b))

6. What is a discrete uniform distribution?
   -  A distribution where all outcomes are equally likely.

      - Example: Rolling a fair die (P(1) = P(2) = … = P(6) = 1/6).

    -  Mean = p, Variance = p(1−p)

       Used to model a single trial.

11. What are the characteristics of a normal distribution?
   -  Bell-shaped, symmetric around mean (μ).

   - Defined by mean (μ) and standard deviation (σ).

   - Total area under curve = 1.
12. What is the standard normal distribution, and why is it important?
   -  A normal distribution with μ = 0 and σ = 1.

  - Used to calculate Z-scores, standardize other normal distributions.
13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?
     -  CLT states: The sampling distribution of the mean approaches a normal distribution as sample size increases (n ≥ 30), regardless of the original distribution.

    - Important for hypothesis testing and confidence intervals.
14. How does the Central Limit Theorem relate to the normal distribution.
    -  CLT justifies why the sample mean follows a normal distribution when sample size is large, even if the population is not normal.
15. What is the application of Z statistics in hypothesis testing?
    -  Z-statistic helps test if a sample mean significantly differs from a population mean.

    - Used in Z-tests when population variance is known or large samples.

17. What are point estimates and interval estimates in statistics?
   -   Point Estimate: Single value estimate (e.g., sample mean).

  - Interval Estimate: Range of values (e.g., confidence interval) that likely includes the true parameter.
18. What is the significance of confidence intervals in statistical analysis?
    -  Gives a range where the true population parameter is likely to lie with a certain confidence level (e.g., 95%).

19. What is the relationship between a Z-score and a confidence interval?
    -   Z-scores define the critical values for confidence intervals.

    -  E.g., 95% CI ⇒ Z = ±1.96
20. How are Z-scores used to compare different distributions
    -   Convert raw scores to standard units, making comparison across different scales possible.
21. What are the assumptions for applying the Central Limit Theorem?
   -   Samples are independent and identically distributed (i.i.d.).

   - Sample size is sufficiently large (typically n ≥ 30).

    - Finite mean and variance.
